# File Write

```python
f = open("count_log.txt", mode = "w", encoding = "UTF8")
for i in range(1, 11):
    data = "{0}번째 줄입니다. \n".format(i)
    f.write(data)
f.close()
```

```python
with open("count_log.txt", mode = "a", encoding = "UTF8") as f:  #'a' mode : append
    for i in range(11, 21):
        data = "{0}번째 줄입니다. \n".format(i)
        f.write(data)
```

```python
import os
if not os.path.isdir("log"): # exists("log")로 써도 됨
    os.mkdir("log")
    
if not os.path.exists("log/count_log.txt"):
    f = open("log/count_log.txt", 'w', encoding="UTF8")
    f.write("기록이 시작됩니다.\n")
    f.close()
```

```python
with open("log/count_log.txt", 'a', encoding="UTF8") as f:
    import random, datetime
    for i in range(1,11):
        stamp = str(datetime.datetime.now())
        value = random.random() * 1000000
        log_line = stamp + "\t" + str(value) + "값이 생성되었습니다." + "\n"
        f.write(log_line)
```

# Pickle

```python
import pickle  # 파이썬의 객체를 영구 저장

f = open("list.pickle", "wb") # binary file로 저장함
test = [1,2,3,4,5]
pickle.dump(test, f)
f.close()
```

```python
del test # test 변수를 지움
```

```python
f = open("list.pickle", "rb") # binary file을 읽어옴
test_pickle = pickle.load(f)
print(test_pickle)  # test 변수를 읽어옴
f.close()
```

```python
import pickle

class Multiply(object):
    def __init__(self, multiplier):
        self.multiplier = multiplier
        
    def multiply(self, number):
        return number * self.multiplier    
```

```python
multiply = Multiply(5)
multiply.multiply(100)
```

```python
f = open("multiply_object.pickle", "wb")
pickle.dump(multiply, f)
f.close()
```

```python
del multiply
```
