# arange

```python
import numpy as np
```

```python
np.arange(30) # list의 range와 같은 효과(0~29까지)
```

```python
np.arange(30).reshape(-1,5)
```

```python
np.arange(0,5,0.5) # floating 단위로도 가능함
```

```python
np.arange(0,5,0.5).tolist()
```

```python
np.zeros((2,5))
```

```python
np.ones(shape=(10,), dtype = np.int8)
```

```python
np.ones((2,5))
```

# something_like
- 기존 ndarray의 shpae 만큼, 1, 0 또는 empty array를 반환 (딥러닝 때 쓰긴 함)

```python
test_matrix = np.arange(30).reshape(5,6)
np.ones_like(test_matrix)
```

# identity
- 단위행렬

```python
np.identity(n=3, dtype=np.int8)
```

```python
np.identity(5)
```

# eye
 - 대각선이 1인 행렬, k값의 시작 index의 변경이 가능

```python
np.eye(N=3, M=5, dtype=np.int8)
```

```python
np.eye(3)
```

```python
np.eye(3,5,k=2) # k= start index
```

# diag
 - 대각행렬의 값을 추출함

```python
matrix = np.arange(9).reshape(3,3)
np.diag(matrix)
```

```python
np.diag(matrix, k=1) # k= start index
```

# random sampling

```python
np.random.uniform(0,1,10).reshape(2,5) # 균등분포
```

```python
np.random.normal(0,1,10).reshape(2,5) # 정규분포
```
